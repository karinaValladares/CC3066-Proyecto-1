### Estrategia de limpieza

- Para las variables DEPARTAMENTO, MUNICIPIO, ESTABLECIMIENTO, NIVEL, SECTOR, AREA, 
  STATUS, MODALIDAD, JORNADA, PLAN, DEPARTAMENTAL tienen pocos datos faltantes por lo que podemos simplemente eliminar los registros dado que son relativamente pocos (menos de 300) comparado con la cantidad total de registros (33, 366).

 - AREA: se modifica su categoría a un código numérico
 
 - STATUS: se modifica su categoría a un código numérico
 
 - MODALIDAD: se modifica su categoría a un código numérico
 
 - JORNADA: se modifica su categoría a un código numérico
 
 - SECTOR: se modifica su categoría a un código numérico
 
 - PLAN: se modifica su categoría a un código numérico
 
 

```python
#creamos una función para las categorías, así se les asigna un valor numérico
def categoria_num(df, categoria, valor_0, valor_1, valor_2, valor_3, valor_4, valor_5, valor_6, valor_7, 
                    valor_8):
    resultado = []
    
    for i in range(0, len(df)):
        if df[categoria][i]== valor_0:
            resultado.append(0)
        elif df[categoria][i] == valor_1:
            resultado.append(1)
        elif df[categoria][i] == valor_2:
            resultado.append(2)
        elif df[categoria][i] == valor_3:
            resultado.append(3)
        elif df[categoria][i] == valor_4:
            resultado.append(4)
        elif df[categoria][i] == valor_5:
            resultado.append(5)  
        elif df[categoria][i] == valor_6:
            resultado.append(6)
        elif df[categoria][i] == valor_7:
            resultado.append(7)
        elif df[categoria][i] == valor_8:
            resultado.append(8)
        else:
            resultado.append(df[categoria][i])
    df[categoria] = resultado
    
    return print(valor_0, ": 0\n", 
                 valor_1, ": 1\n",
                valor_2, ": 2\n",
                valor_3, ": 3\n",
                valor_4, ": 4\n",
                valor_5, ": 5\n",
                 valor_6, ": 6\n",
                 valor_7, ": 7\n",
                 valor_8, ": 8\n")
```
